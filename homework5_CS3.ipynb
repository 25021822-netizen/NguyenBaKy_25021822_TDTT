```python
# W6A16
n, m = map(int, input().split())
a = []
for _ in range(n):
    a.append(list(map(int, input().split())))
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:>4}", end="")
    print()
```

```python
# W6A19
def max_pos(lst):
    return lst.index(max(lst))

a = list(map(int, input().split()))
print(max_pos(a))
```

```python
# W6A20
def find_k(lst, k):
    if k in lst:
        return lst.index(k)
    return -1

a = list(map(int, input().split()))
k = int(input())
print(find_k(a, k))
```

```python
# W6A21
s = input().split()
d = {}

for i in range(len(s)):
    if s[i] not in d:
        d[s[i]] = []
    d[s[i]].append(i)

for k in d:
    d[k] = tuple(d[k])

print(d)
    
```

```python
#6A23
def ham(s, k):
    a = []
    for x in s:
        if x >= k:
            a.append(x)
    return a
if __name__ == '__main__':
    s = list(map(int, input().split()))
    x = int(input())
    print(ham(s, x))
```

```python
# W6A28
a = list(map(int, input().split()))

tang = True
giam = True

for i in range(1, len(a)):
    if a[i] <= a[i-1]:
        tang = False
    if a[i] >= a[i-1]:
        giam = False

if tang:
    print("tang thuc su")
elif giam:
    print("giam thuc su")
else:
    print("vo danh")
```
