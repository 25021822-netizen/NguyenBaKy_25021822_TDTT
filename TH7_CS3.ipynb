```python
#7w1
def timkiem(a, l, r, x):
    while l <= r:
        m = (l+r)//2
        if a[m] == x:
            return True, m
        elif a[m]>x:
            r = m-1
        elif a[m] < x:
            l = m+1
    return False, -1
a = list(map(int, input().split()))
l, r = 0, len(a)-1
x = int(input())
w, z = timkiem(a, l, r, x)
print(z)
```

```python
# W7A7
danh_sach = list(map(int, input().split()))
so_luong = len(danh_sach)

do_dai = [1] * so_luong

for i in range(so_luong):
    for j in range(i):
        if danh_sach[j] < danh_sach[i]:
            do_dai[i] = max(do_dai[i], do_dai[j] + 1)

print(max(do_dai))
```

```python
# W7A8
cac_khoang = eval(input())
cac_truy_van = list(map(int, input().split()))

ket_qua = []
for gia_tri in cac_truy_van:
    do_dai_nho_nhat = 10**18
    for trai, phai in cac_khoang:
        if trai <= gia_tri <= phai:
            do_dai_nho_nhat = min(do_dai_nho_nhat, phai - trai + 1)
    if do_dai_nho_nhat == 10**18:
        ket_qua.append(-1)
    else:
        ket_qua.append(do_dai_nho_nhat)

print(ket_qua)
```

```python
# W7A10
from bisect import bisect_right, insort
a = eval(input())
m = int(input())
pre = 0
ans = 0
s = []
for x in a:
    pre = (pre + x) % m
    ans = max(ans, pre)
    i = bisect_right(s, pre)
    if i < len(s):
        ans = max(ans, (pre - s[i] + m) % m)
    insort(s, pre)
print(ans)
```

```python
# W7A13
from collections import Counter
danh_sach = list(map(int, input().split()))
tan_suat = Counter(danh_sach)
do_dai_lon_nhat = 0
for gia_tri in tan_suat:
    if gia_tri + 1 in tan_suat:
        do_dai_lon_nhat = max(
            do_dai_lon_nhat,
            tan_suat[gia_tri] + tan_suat[gia_tri + 1]
        )
print(do_dai_lon_nhat)
```
