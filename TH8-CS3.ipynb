```python
# W7A16
from collections import Counter

a = list(map(int, input().split()))
c = Counter(a)

del_cnt = 0
ok = True
for x in c:
    if c[x] != x:
        ok = False
        if c[x] > x:
            del_cnt += c[x] - x
        else:
            del_cnt += c[x]

if ok:
    print("YES")
else:
    print("NO", del_cnt)
```

```python
# W7A17
a = list(map(int, input().split()))

cand = None
cnt = 0
for x in a:
    if cnt == 0:
        cand = x
        cnt = 1
    elif x == cand:
        cnt += 1
    else:
        cnt -= 1

if a.count(cand) > len(a)//2:
    print(cand)
```

```python
# W7A19
a = list(map(int, input().split()))
b = list(map(int, input().split()))

res = []
i = j = 0
inc = a[0] <= a[-1]

while i < len(a) and j < len(b):
    if (a[i] <= b[j]) == inc:
        res.append(a[i]); i += 1
    else:
        res.append(b[j]); j += 1

res += a[i:] + b[j:]
print(*res)
```

```python
# W7A22
n = int(input())
cnt = {}

for _ in range(n):
    name = input().split()[-1]
    cnt[name] = cnt.get(name, 0) + 1

mx = max(cnt.values())
ans = min(x for x in cnt if cnt[x] == mx)
print(ans)
```

```python
# W7A25
s = input()
k = int(input())

res = []
drop = len(s) - k

for ch in s:
    while drop and res and res[-1] < ch:
        res.pop()
        drop -= 1
    res.append(ch)

print("".join(res[:k]))
```

```python
# W7A27
s = input()
st = []
mp = {')':'(', ']':'[', '}':'{', '>':'<'}

ok = True
for ch in s:
    if ch in "([{<":
        st.append(ch)
    else:
        if not st or st[-1] != mp[ch]:
            ok = False
            break
        st.pop()

print("YES" if ok and not st else "NO")
```

```python
# W7A28
N = int(input())

def comp(x):
    if x < 4: return False
    for i in range(2, int(x**0.5)+1):
        if x % i == 0:
            return True
    return False

best = []
cur = []

for i in range(4, N+1):
    if comp(i):
        cur.append(i)
        if len(cur) > len(best):
            best = cur[:]
    else:
        cur = []

print(*best)
```

```python
# W7A29
a = list(map(int, input().split()))

best = cur = [a[0]]
for i in range(1, len(a)):
    if a[i] > a[i-1]:
        cur.append(a[i])
    else:
        if len(cur) > len(best):
            best = cur
        cur = [a[i]]

if len(cur) > len(best):
    best = cur

print(*best)
```

```python
# W7A30
a = list(map(int, input().split()))
n = len(a)

dp = [[] for _ in range(n)]
for i in range(n):
    dp[i] = [a[i]]
    for j in range(i):
        if a[j] < a[i] and len(dp[j]) + 1 > len(dp[i]):
            dp[i] = dp[j] + [a[i]]

ans = max(dp, key=len)
print(*ans)
```

```python
# W7A31
s = list(input())
n = len(s)
best = s[:]

for i in range(n):
    for j in range(i+1, n):
        s[i], s[j] = s[j], s[i]
        if s[0] != '0' and s < best:
            best = s[:]
        s[i], s[j] = s[j], s[i]

print("".join(best))
```
