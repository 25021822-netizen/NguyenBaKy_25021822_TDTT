```python
#4W8
def reverse_num(x):
    r = 0
    while x > 0:
        r = r * 10 + x % 10
        x //= 10
    return r

def is_palindrome(x):
    return x == reverse_num(x)

n = int(input())
steps = 0

while not is_palindrome(n):
    n = n + reverse_num(n)
    steps += 1

print(steps, n)

```

```python
#4w9
from math import*
n = int(input())
def kt(x):
    cnt = [0] * 10
    while x:
        t = x% 10
        cnt[t] += 1
        x //=10
        if cnt[t] >1:
            return False
    return True
for i in range(1, isqrt(n)):
    if kt(i*i):
        print(i*i, end = ' ')
```
