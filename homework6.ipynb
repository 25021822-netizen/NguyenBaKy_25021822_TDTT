```python
#6w19
def max_val(l):
    return l.index(max(l))
l = list(map(int, input().split()))
print(max_val(l))
```

```python
#6w20
def f(l, k):
    for i in range(len(l)):
        if l[i] == k:
            return i
    return -1
if __name__ == '__main__':
    l = list(map(int, input().split()))
    k = int(input())
    print(f(l, k))
            
```

```python
#6w21
l = input().split()
d = {}
for x in l:
    if x not in d.keys():
        d[x] = 0

print(d)
```
