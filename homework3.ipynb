```python
# bài 20
def nt(n):
    if n < 2:
        return False
    else:
        for i in range(2, isqrt(n) + 1):
            if n % i == 0:
                return False
        return True


n = int(input())
dem = 0
while n != 0:
    a = n % 10
    if nt(a):
        dem += 1
        n //= 10
print(dem)
    
```
