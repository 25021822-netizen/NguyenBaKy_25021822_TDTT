```python
#bai 3
from math import*
x = int(input())
def ktsnt(x):
    if x < 2 :
        return False
    for i in range(2, isqrt(x)+1):
            if x % i == 0:
                return False
    return True

if ktsnt(x):
    print(x)



```

```python
#5W7
n1, n2 = map(int, input().split())
phep = input()
def maytinh(n1, n2, phep):
    if phep == '+':
        return n1 + n2
    elif phep == '-':
        return n1 - n2
    elif phep == '*':
        return n1 * n2
    else:
        return n1 / n2

print('%.2f' % (maytinh(n1, n2, phep)))
```

```python
#5W8
x, y = map(int, input().split())
x, y = bin(x)[2:], bin(y)[2:]
if len(x) > len(y) :
    m = len(x)- len(y)
    y = '0'*m + y
elif len(y) > len(x) :
    m = len(y)- len(x)
    x = '0'*m + x
    
def sosanh(x, y):
    n = min(len(x), len(y))
    dem = 0
    for i in range(n):
        if x[i] != y[i]:
            dem+=1

    dem = dem + abs(len(x)-len(y))
    return dem

print(sosanh(x, y))
        

```

```python
#5W9
n = int(input())
def tong(x):
    tong = 0
    for i in range(1, 1000):
        tong += 1
        x //= 10
        if x == 0:
            break

    return tong

print(tong(n))
```

```python
#5W10
x, y = input().split()
def anhxa(x, y):
    m = len(x)
    for i in range(m):
        if x[i] == y[i]:
            return True
        else:
            return False

print(anhxa(x, y))
    
```

```python
# bài tập luyện thêm
# số Amstrong
n = int(input())
x = str(n)
k = len(x)
m = n
tong = 0
while n != 0:
    tong += (n % 10) ** k
    n //= 10
    
def amstrong(n):
    if tong == m :
        return True
    else:
        return False
print(tong)
print(amstrong(n))
    
```

```python
#số đối xứng
n = int(input())
m=n
def sodx(n):
    somoi = 0
    while n != 0:
        somoi = somoi*10 + (n % 10)
        n //= 10
    if somoi == m :
        return True
    else:
        return False

print(sodx(n))
```

```python
#số kỳ diệu
n = int(input())
def skd(x):
    while x >= 10:
        tong = 0
        while x != 0:
            tong += x % 10
            x //= 10
        x = tong
    if tong == 1:
        return True
    else:
        return False
print(skd(n))
```
