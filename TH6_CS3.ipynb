```python
#6w4
n = input().split()
d = {}
for x in n:
    k, v = x.split(':')
    if k not in d:
        d[k] = []
    d[k].append(v)
print(d)
```

```python
#6w5
a = list(map(int, input().split()))
d = {'positives':0, 'negatives':0, 'zeros':0}
for x in a:
    if x > 0:
        d['positives'] += 1
    if x < 0:
        d[ 'negatives'] += 1
    if x == 0:
        d['zeros'] += 1
print(d)
```

```python
#6W9
d1 = {}
d2 = {}
n1 = input().split()
n2 = input().split()
for x in n1:
    k, v = x.split(':')
    d1[k] = int(v)
for x in n2:
    k, v = x.split(':')
    d2[k] = int(v)
for k, v in d2.items():
    d1[k] = d1.get(k, 0) + v
key = sorted(d1)
res = {}
for k in key:
    res[k] = d1[k]
print(res)
```

```python
#6w12
a = list(map(int, input().split()))
d = {}
res = 10**18
max_val = -10**18
for x in a:
    d[x] = d.get(x, 0) + 1
for k, v in d.items():
    if v >= max_val:
        max_val = v
        if k < res:
            res = k
print(res)

    
```

```python
#6w14
n1 = list(map(int, input().split()))
n2 = list(map(int, input().split()))
a = set(n2)
res = []
for x in n1:
    for y in a:
        if x == y:
            if x not in res:
                res.append(x)
print(res)
            
```
