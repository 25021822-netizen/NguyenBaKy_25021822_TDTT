```python
# W9A1
a = list(map(int, input().split()))
b = list(map(int, input().split()))

i = j = 0
res = []

while i < len(a) and j < len(b):
    if a[i] <= b[j]:
        res.append(a[i]); i += 1
    else:
        res.append(b[j]); j += 1

res += a[i:] + b[j:]
print(*res)
```

```python
# W9A2
n, m = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]
k = int(input())

lo, hi = a[0][0], a[-1][-1]

while lo < hi:
    mid = (lo + hi + 1) // 2
    cnt = 0
    j = m - 1
    for i in range(n):
        while j >= 0 and a[i][j] > mid:
            j -= 1
        cnt += j + 1
    if cnt >= k:
        lo = mid
    else:
        hi = mid - 1

print(lo)
```

```python
# W9A3
s = list(input())
n = len(s)

best = s[:]
for i in range(n):
    for j in range(i+1, n):
        s[i], s[j] = s[j], s[i]
        if s[0] != '0' and s < best:
            best = s[:]
        s[i], s[j] = s[j], s[i]

print("".join(best))
```

```python
# W9A4
q = int(input())

for _ in range(q):
    a = input()
    b = input()
    i = j = 0

    while i < len(a) and j < len(b):
        if a[i].upper() == b[j]:
            j += 1
        elif a[i].isupper():
            break
        i += 1

    ok = (j == len(b))
    for k in range(i, len(a)):
        if a[k].isupper():
            ok = False

    print("YES" if ok else "NO")
```

```python
# W9A5
a = list(map(int, input().split()))
b = list(map(int, input().split()))

ok = False
for i in range(len(b) - len(a) + 1):
    if b[i:i+len(a)] == a:
        ok = True
        break

print("YES" if ok else "NO")
```

```python
# W9A6
n = int(input())
a = list(map(int, input().split()))

i = n - 2
while i >= 0 and a[i] > a[i+1]:
    i -= 1

if i < 0:
    a.reverse()
else:
    j = n - 1
    while a[j] < a[i]:
        j -= 1
    a[i], a[j] = a[j], a[i]
    a[i+1:] = reversed(a[i+1:])

print(*a)
```

```python
# W9A7
bed = list(map(int, input().split()))
k = int(input())

i = 0
while i < len(bed) and k > 0:
    if bed[i] == 0:
        l = (i == 0 or bed[i-1] == 0)
        r = (i == len(bed)-1 or bed[i+1] == 0)
        if l and r:
            bed[i] = 1
            k -= 1
    i += 1

print(k == 0)
```

```python
# W9A8
s = input()

x = y = 0
vis = {(0,0)}

ok = False
for c in s:
    if c == 'R': x += 1
    if c == 'L': x -= 1
    if c == 'U': y += 1
    if c == 'D': y -= 1
    if (x,y) in vis:
        ok = True
        break
    vis.add((x,y))

print(ok)
```

```python
# W9A9 (ít so sánh)
a = list(map(int, input().split()))
x1, x2, x3, x4, x5 = a

if x1 > x2: x1, x2 = x2, x1
if x3 > x4: x3, x4 = x4, x3
if x1 > x3: x1, x3 = x3, x1
if x2 > x4: x2, x4 = x4, x2
if x2 > x3: x2, x3 = x3, x2

if x5 < x2:
    print(x2)
elif x5 > x3:
    print(x3)
else:
    print(x5)
```

```python
# W9A10
def findRadius(h, t):
    h.sort()
    t.sort()
    i = 0
    ans = 0

    for x in h:
        while i+1 < len(t) and abs(t[i+1] - x) <= abs(t[i] - x):
            i += 1
        ans = max(ans, abs(t[i] - x))
    return ans


houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
